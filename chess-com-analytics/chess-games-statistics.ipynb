```python
import pandas as pd
from pathlib import Path
import re
import matplotlib.pyplot as plt


#load data
df = pd.read_csv("all_chess_games.csv")
```

```python
# Example: Count wins/losses/draws by weekday
df.groupby(["Weekday", "My Result"]).size().unstack().plot(kind="bar", figsize=(10,6))
```

```python
#win to loss ratio
win_loss_counts = df[df["My Result"].isin(["Win", "Loss"])].groupby(["Weekday", "My Result"]).size().unstack(fill_value=0)
win_loss_ratio = (win_loss_counts["Win"] / win_loss_counts["Loss"]).sort_index()

# Define correct weekday order
weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

# Reindex to correct order before plotting
win_loss_ratio = win_loss_ratio.reindex(weekday_order)

# Plot
win_loss_ratio.plot(kind="bar", figsize=(10, 6), title="Win/Loss Ratio per Weekday", ylabel="Win/Loss Ratio", xlabel="Weekday")
```

```python
#win loss ration per color
# Filter only bullet games (TimeControl == '60') and results of interest
bullet_games = df[(df["TimeControl"] == "60") & (df["My Result"].isin(["Win", "Loss"]))]

# Group by Weekday, Color, and Result
grouped_bullet = (
    bullet_games
    .groupby(["Weekday", "Player Color Me", "My Result"])
    .size()
    .unstack(fill_value=0)
)

# Compute Win/Loss ratio
grouped_bullet["Win/Loss Ratio"] = grouped_bullet["Win"] / grouped_bullet["Loss"]

# Pivot for plotting
pivot_bullet = grouped_bullet["Win/Loss Ratio"].unstack("Player Color Me")

# Reorder weekdays
weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
pivot_bullet = pivot_bullet.reindex(weekday_order)



# Plot setup
ax = pivot_bullet.plot(
    kind="bar",
    figsize=(10, 6),
    edgecolor='black',
    title="Bullet Games: Win/Loss Ratio per Weekday by Color",
    ylabel="Win/Loss Ratio",
    xlabel="Weekday",
    color={"White": "white", "Black": "black"}
)

# Improve visibility of white bars with black edges
for bar in ax.patches:
    if bar.get_facecolor() == (1.0, 1.0, 1.0, 1.0):  # white bars
        bar.set_edgecolor("black")
```

```python
# Group bullet games by Weekday and Result (ignoring color)
grouped_combined = (
    bullet_games
    .groupby(["Weekday", "My Result"])
    .size()
    .unstack(fill_value=0)
)

# Compute win/loss ratio
grouped_combined["Win/Loss Ratio"] = grouped_combined["Win"] / grouped_combined["Loss"]

# Reindex to correct weekday order
weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
grouped_combined = grouped_combined.reindex(weekday_order)

# Plot
grouped_combined["Win/Loss Ratio"].plot(
    kind="bar",
    figsize=(10, 6),
    color="gray",
    edgecolor="black",
    title="Bullet Games: Win/Loss Ratio per Weekday (Combined Colors)",
    ylabel="Win/Loss Ratio",
    xlabel="Weekday"
)
```
